## PCA Graph

```python
def compute_scores(X,n_components):
    pca = PCA(svd_solver='full')

    pca_scores = []
    for n in range(1,n_components):
        pca.n_components = n
        pca_scores.append(np.mean(cross_val_score(pca, X, cv=5)))
        
    return pca_scores
```

```python
def clear(X1,X2,X3):
    attack=pd.concat((X3.iloc[:,:5],X1.iloc[:,1],X2['Offsides pg']),axis=1)
    defence=X2.iloc[:,:5]
    control=X1.iloc[:,[0,4,5,6]]
    attack1=normalize(attack.iloc[:,1:], norm='l2')
    defence1=normalize(defence.iloc[:,1:], norm='l2')
    control1=normalize(control.iloc[:,1:], norm='l2')
    return(attack1,defence1,control1)
```

```python
def component(x,y,z):
    p1=PCA(3).fit_transform(x)
    p2=PCA(2).fit_transform(y)
    p3=PCA(1).fit_transform(z)
    return(p1,p2,p3)
    
    
```

```python
attack1,defence1,control1=clear(X1,X2,X3)
```

```python
p1,p2,p3=component(attack1,defence1,control1)
```

```python
cluster=KMeans(4,random_state=0).fit(p1)
cluster.labels_
```

```python
cluster2=KMeans(4,random_state=0).fit(p2)
cluster2.labels_
```

```python
cluster3=KMeans(4,random_state=0).fit(p3)
cluster3.labels_
```

```python
def ratio(X,n):
    ratio=[]
    for i in range(1,n+1):
            pca=PCA(i)
            pca.fit(X)
            ratio.append(pca.explained_variance_ratio_.sum())
    return ratio
```

```python
def confusion_matrix_multiclass(y_true, y_predicted):
    num = np.unique(y_true).shape[0]
    matrix=np.zeros((num,num)).astype(int)
    for i,j in zip(y_true,y_predicted):
        matrix[i+1,j+1]+=1
    return matrix
```
